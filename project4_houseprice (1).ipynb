```python
from scipy import stats
```

```python
def diagostic_plot(df, variable):
    plt.figure(figsize=(15,6))
    plt.subplot(1,2,1)
    df[variable].hist()


    plt.subplot(1,2,2)
    stats.probplot(df[variable],dist='norm',plot=plt)
    plt.show()
```

```python
diagostic_plot(df,'SQUARE_FT')
```

```python
df['lat']=np.log(df['SQUARE_FT']+1)
diagostic_plot(df,'lat')
```

```python
df['ref']= 1/(df['SQUARE_FT']+1)
diagostic_plot(df,'ref')
```

```python
df['squre']=df['SQUARE_FT']**1/2
diagostic_plot(df,'squre')
```

```python
df['exp_ft']=df['SQUARE_FT']**1/5
diagostic_plot(df,'exp_ft')
```
